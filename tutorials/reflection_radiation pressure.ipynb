```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
import scienceplots
import matplotlib as mpl
```

```python
lambdas = np.array([2, 1, 0.5, 1/5, 1/10, 1/20])
lambdas_2 = np.array([2, 1, 0.5, 1/5, 1/10, 1/25])

#F_y_square = np.array([4.8, 5.43, 4.46, 5.31, 5.34, 5.34])
F_y_square = np.array([4.54, 5.22, 5.4, 5.3, 5.33, 5.32])
#F_y_square = np.array([1.85, 2.34, 2.62, 2.64, 2.66, 2.67])
F_y_dom = np.array([31.25, 33.50, 35.10, 36.36, 37.02, 37.32])
```

```python
F_t_square = 6.2
F_t_dom = 37.54
```

```python
with plt.style.context(['science', 'nature']):

    fig, ax = plt.subplots(figsize=(4,2))
    ax.scatter(lambdas_2, F_y_square, color="blue", label ="Data points", alpha=1, edgecolors='black', linewidths=0.5, s=45)
    ax.plot(lambdas_2, F_y_square, color="blue", alpha=0.5, linewidth=0.5, linestyle="dashed")
    ax.axhline(y=F_t_square, color='green', linestyle='--', label="bound")


    ax.set_xlabel("$\\lambda$", fontsize=14)
    ax.set_ylabel("$\\langle F_y \\rangle $(pN/\\textmu m)", fontsize=14)
    ax.set_xscale("log")
    ax.legend(frameon=True, fontsize=12)
    plt.gca().invert_xaxis()
    ax.tick_params(axis='both', which='major', labelsize=10)
fig
```
